Across all three models, structured agentic navigation substantially outperforms random-walk
exploration on every distance and accuracy metric, confirming that the LLM reasoning loop
contributes beyond raw visual coverage. GPT-5-mini sees the largest absolute gain, improving
from 1521.70 km under random-walk to 899.99 km under agentic navigation, alongside consistent
accuracy gains at every geographic granularity. Gemma-3-27B shows the most dramatic separation,
with agentic navigation reducing mean distance by over 2,600 km relative to random-walk. Notably,
reasoning traces are longer in the random-walk condition across all three models, yet performance
is worse, reinforcing the cross-model finding that verbosity alone does not drive accuracy and
that the quality of goal-directed reasoning matters more than its length.
